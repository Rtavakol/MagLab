```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = 108 # Numbers of wires per quadrant, Even number
L = 27 # Lemgth of the coil
x = np.linspace(-L/2, L/2, N)
y = np.linspace(-L/2, L/2, N)
```

```python
# 1st Quadrant:
init, end = 1, -1
L1_top_Q1 = [(i, -L/2, L) for i in y[init:end]]
L2_top_Q1 = [(i, -np.abs(j), L) for i,j in zip(x[init:end], y[init:end])]
L1_bottom_Q1 = [(i, -L/2, 0) for i in y[init:end]]
L2_bottom_Q1 = [(i, -np.abs(j), 0) for i,j in zip(x[init:end], y[init:end])]

# 2nd Quadrant
L1_top_Q2 = [(L/2, i, L) for i in y[init:end]]
L2_top_Q2 = [(np.abs(i), j, L) for i,j in zip(x[init:end], y[init:end])]
L1_bottom_Q2 = [(L/2, i, 0) for i in y[init:end]]
L2_bottom_Q2 = [(np.abs(i), j, 0) for i,j in zip(x[init:end], y[init:end])]


# 3rd Quadrant
L1_top_Q3 = [(-i, L/2, L) for i in y[init:end]]
L2_top_Q3 = [(-i, np.abs(j), L) for i,j in zip(x[init:end], y[init:end])]
L1_bottom_Q3 = [(-i, L/2, 0) for i in y[init:end]]
L2_bottom_Q3 = [(-i, np.abs(j), 0) for i,j in zip(x[init:end], y[init:end])]


# 4th Quadrant
L1_top_Q4 = [(-L/2, -i, L) for i in y[init:end]]
L2_top_Q4 = [(-np.abs(i), -j, L) for i,j in zip(x[init:end], y[init:end])]
L1_bottom_Q4 = [(-L/2, -i, 0) for i in y[init:end]]
L2_bottom_Q4 = [(-np.abs(i), -j, 0) for i,j in zip(x[init:end], y[init:end])]
```

```python
%matplotlib notebook
plt.plot(x[1:-1], -np.abs(y[1:-1]), marker = 'o')
plt.plot(x[1:-1], [-L/2 for i in y[1:-1]], marker = 'o')
```

```python
def print_start_point(p):
    print('start    {:.1f}    {:.1f}    {:.1f}   cm'.format(*p))
    
def print_coordinates(p):
    print('goto    {:.1f}    {:.1f}    {:.1f}    cm'.format(*p))
```

```python
def First_Quadrant():
    for i in range(N-1):
        # 1st Quadrant
        if i == 0:
            print_coordinates(L1_top_Q1[i])
        if i < N - 3 :  
            print_coordinates(L2_top_Q1[i])  
            print_coordinates(L2_bottom_Q1[i+1])
            print_coordinates(L1_bottom_Q1[i+1])
            print_coordinates(L1_top_Q1[i + 1])
        if i == N - 3:
            print_coordinates(L2_top_Q1[i])
    
    print('\n \n \n \n')
    # Connection to the next quadrant
    print('goto    {:.1f}    {:.1f}    {:.1f}    cm'.format(L/2, -L/2, L/2))
    print_coordinates(L1_bottom_Q2[0])
    print('\n \n \n \n')
```

```python
def Second_Quadrant():
    for i in range(N-1):
        # 2nd Quadrant
        if i == 0:
            print_coordinates(L1_top_Q2[i])
        if i < N - 3 :  
            print_coordinates(L2_top_Q2[i])  
            print_coordinates(L2_bottom_Q2[i+1])
            print_coordinates(L1_bottom_Q2[i+1])
            print_coordinates(L1_top_Q2[i + 1])
        if i == N - 3:
            print_coordinates(L2_top_Q2[i])
    print('\n \n \n \n')
    # Connection to the next quadrant
    print('goto    {:.1f}    {:.1f}    {:.1f}    cm'.format(L/2, L/2, L/2))
    print_coordinates(L1_bottom_Q3[0])
    print('\n \n \n \n')
```

```python
def Third_Quadrant():
    for i in range(N-1):
        # 3rd Quadrant
        if i == 0:
            print_coordinates(L1_top_Q3[i])
        if i < N - 3 :  
            print_coordinates(L2_top_Q3[i])  
            print_coordinates(L2_bottom_Q3[i+1])
            print_coordinates(L1_bottom_Q3[i+1])
            print_coordinates(L1_top_Q3[i + 1])
        if i == N - 3:
            print_coordinates(L2_top_Q3[i])
    print('\n \n \n \n')
    # Connection to the next quadrant
    print('goto    {:.1f}    {:.1f}    {:.1f}    cm'.format(-L/2, L/2, L/2))
    print_coordinates(L1_bottom_Q4[0])
    print('\n \n \n \n')
```

```python
def Fourth_Quadrant():
    for i in range(N-1):
        # 3rd Quadrant
        if i == 0:
            print_coordinates(L1_top_Q4[i])
        if i < N - 3 :  
            print_coordinates(L2_top_Q4[i])  
            print_coordinates(L2_bottom_Q4[i+1])
            print_coordinates(L1_bottom_Q4[i+1])
            print_coordinates(L1_top_Q4[i + 1])
        if i == N - 3:
            print_coordinates(L2_top_Q4[i])
    print('\n \n \n \n')
    # Connection to the next quadrant
    print('goto    {:.1f}    {:.1f}    {:.1f}    cm'.format(-L/2, -L/2, L/2))
    print_coordinates(L1_bottom_Q1[0])
    print_coordinates(L1_top_Q1[0])
    print('\n \n \n \n')
```

### Magnetic Field Visualization

```python
df = pd.read_csv('Bx-Contour-plot.txt', sep = '\t', header = None)
df.columns = ['X (cm)', 'Y (cm)', 'Z (cm)', 'Bx (T)', 'By (T)', 'Bz (T)', 'B (T)']
df.shape
```

```python
df = df[df['B (T)'] < 4e-6]
```

```python
df.head()
```

```python
X = df['X (cm)']
Y = df['Y (cm)']
U = df['Bx (T)']
V = df['By (T)']
```

```python
fig, ax = plt.subplots(figsize = (6,6))
d = 1
#ax.set_title('Arrows scale with plot width, not view')
Q = ax.quiver(X[::d], Y[::d], U[::d], V[::d], units='width', pivot = 'middle')
qk = ax.quiverkey(Q, 0.5, 0.9, 2e-6, r'$2 \mu T$', labelpos='E',
                   coordinates='figure', color = 'r')
```
