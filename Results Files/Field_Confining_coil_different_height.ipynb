```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import glob as glob
```

```python
files = glob.glob('B_vs_x_Height_*_cm.txt')
files
```

```python
%matplotlib notebook
fig, axe = plt.subplots(2, figsize = (6, 6))
for file in files:
    
    df_B = pd.read_csv(file, sep = '\t', header = None)
    df_B.columns = ['X (m)', 'Y (m)', 'Z (m)', 'Bx (T)', 'By (T)', 'Bz (T)', 'B (T)']
    df_B = df_B[df_B['B (T)'] < 1e4]
    df_B.head()
    
    axe[0].plot(df_B['X (m)'], df_B['By (T)']*1e6, label = file[7:-4])
    axe[1].plot(df_B['X (m)'], np.abs(df_B['By (T)']*1e6), label = file[7:-4])
    

    axe[0].legend()
    axe[1].legend()
axe[1].set_ylim(1.17,1.19)
```

```python
files_z = glob.glob('B_vs_z_Height_*_cm.txt')
files_z
```

```python
%matplotlib notebook
fig, axe = plt.subplots(2, figsize = (6, 6))
for file in files_z:
    
    df_B = pd.read_csv(file, sep = '\t', header = None)
    df_B.columns = ['X (m)', 'Y (m)', 'Z (m)', 'Bx (T)', 'By (T)', 'Bz (T)', 'B (T)']
    df_B = df_B[df_B['B (T)'] < 1e4]
    df_B.head()
    
    axe[0].plot(df_B['Z (m)'], df_B['By (T)']*1e6, label = file[7:-4])
    axe[1].plot(df_B['Z (m)'], np.abs(df_B['By (T)']*1e6), label = file[7:-4])
    

    axe[0].legend()
    axe[1].legend()
    
axe[0].set_xlabel('Z (cm)')
axe[1].set_xlabel('Z (cm)')
axe[0].set_ylabel(r'$B_y$ ($\mu T$)')
axe[1].set_ylabel(r'$B_y$ ($\mu T$)')

axe[1].set_ylim(1.17,1.174)
```
